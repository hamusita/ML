```python
df1 = pd.read_csv("winequality-red.csv", sep=';')
df2 = pd.read_csv("winequality-white.csv", sep=';')
```

```python
df1.head()
```

```python
df2.head()
```

```python
wine_features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']
```

```python
df1.describe()
```

```python
df2.describe()
```

```python
df1['color'] = "red"
df2["color"] = "white"
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.concat([df1, df2])
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(df, hue="color")
plt.show()
```

```python
x, y = df[wine_features+["quality"]], df["color"]
```

```python
from sklearn.model_selection import train_test_split
x_tr, x_te, y_tr, y_te = train_test_split(x, y, train_size=0.7, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

clf = DecisionTreeClassifier(random_state=1, criterion='entropy')
clf.fit(x_tr, y_tr)
print("test score : {:.3f}".format(clf.score(x_te, y_te)))
plt.figure(figsize=(30, 15))
plot_tree(clf, filled=True)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_tr, y_tr)
print("test score : {:.3f}".format(knn.score(x_te, y_te)))
```

```python
X, Y = df[wine_features], df["quality"]
```

```python
X_tr, X_te, Y_tr, Y_te = train_test_split(X, Y, train_size=0.7, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor 
from sklearn.metrics import mean_absolute_error

clf = DecisionTreeRegressor (random_state=1)
clf.fit(X_tr, Y_tr)
wine_preds = clf.predict(X_te)
print("MAE : {:.3f}".format(mean_absolute_error(Y_te, wine_preds)))
plt.figure(figsize=(30, 15))
plot_tree(clf, filled=True)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=1)
knn.fit(X_tr, Y_tr)
wine_preds = knn.predict(X_te)
print("MAE : {:.3f}".format(mean_absolute_error(Y_te, wine_preds)))
```

# 1.はじめに
このノートブックはUCIが提供するワインのクオリティーについて、解析と機械学習による推論を行った記録を記したものである。以下にその詳細を記す。


# 2.手法
今回は[UCIが提供するワインのデータセット](https://archive.ics.uci.edu/ml/datasets/Wine+Quality)を用いてワインが白ワインか赤ワインかを推論する分類モデルと、そのワインのクオリティーを推論する回帰モデルを複数作成し比較を行った。
## 2.1用いた手法
今回の学習では2つのアルゴリズムを用いてそれぞれ分類と回帰モデルを作成し学習した。

使用したモデルは以下の通りである。
### 2.1.1 決定木
学習モデルの1つ目として授業中に習った決定木を用いた。

モデル選定の理由として、決定木の可視化により推論のプロセスを確認しやすいことが挙げられる。

また、決定木の作成に用いる指標はエントロピーとした。

### 2.1.2 k-近傍法
学習モデルの2つ目としてk-近傍法を用いた。

このモデルは、機械学習モデルの中で最も簡単といってよいほど簡単なアルゴリズムで、同じくプロセスの確認が容易なため用いた。

今回は簡単のためk=1とし最近傍探索とした。

## 2.2結果
赤ワインと白ワインの分類では決定木での正解率が98.9%、近傍法での正解率が95.4%となり決定木での推論のほうが優れた結果となった。

ワインのクオリティー推測では決定木の誤差が0.534、近傍法での誤差が0.566となりわずかながら決定木のほうが優れた結果となった。

ただし、ここで用いた誤差は平均絶対誤差(MAE)とした。

# 3.考察
2つのモデルを用いて比較を行った結果、2.2のようになった。

まず分類について考える。

分類では決定木のほうが4.5%正解率が高く、近傍法より優れた結果となった。

決定木を図示したものを散布図行列と比較しながら見てみると、当然ではあるがしっかりと情報量の大小に従い決定木が作成されていることが分かる。



次に回帰について考える。

回帰でも決定木のほうが誤差が0.034小さいが、差が小さいため決定木のほうが有意に優れているとは言い難いだろう。

また、データセットの統計量を見てみると、クオリティーの平均値は5.818である。

よってこれらのモデルでの推測値は平均の10%近い値となることがわかり、この推論モデルでの推論はとてもではないが信頼性が高いとは言えないと考えられる。
